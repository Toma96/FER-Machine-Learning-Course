U prvoj laboratorijskoj vježbi koristili smo model linearne regresije za, naravno, regresiju. Međutim, model linearne regresije može se koristiti i za **klasifikaciju**. Iako zvuči pomalo kontraintuitivno, zapravo je dosta jednostavno. ### 3. Logistička regresija

Ovaj zadatak bavi se probabilističkim diskriminativnim modelom, **logističkom regresijom**, koja je, unatoč nazivu, klasifikacijski model.

Logistička regresija tipičan je predstavnik tzv. **poopćenih linearnih modela** koji su oblika: $h(\mathbf{x})=f(\mathbf{w}^\intercal\tilde{\mathbf{x}})$. Logistička funkcija za funkciju $f$ koristi tzv. **logističku** (sigmoidalnu) funkciju $\sigma (x) = \frac{1}{1 + \textit{exp}(-x)}$.

#### (a)

Definirajte logističku (sigmoidalnu) funkciju $\mathrm{sigm}(x)=\frac{1}{1+\exp(-\alpha x)}$ i prikažite je za $\alpha\in\{1,2,4\}$.

```python
# Vaš kôd ovdje
def sigmoid(x, alpha=1):
    return 1 / (1 + np.exp(-alpha * x))

x = np.arange(-5, 5, 0.01)
for alpha in [1, 2, 4]:
    y = sigmoid(x, alpha=alpha)
    plt.plot(x, y)
plt.legend(loc="best", labels=[1, 2, 4])
```

#### (b)

Implementirajte funkciju 

> `lr_train(X, y, eta=0.01, max_iter=2000, alpha=0, epsilon=0.0001, trace=False)` 

za treniranje modela logističke regresije gradijentnim spustom (*batch* izvedba). Funkcija uzima označeni skup primjera za učenje (matrica primjera `X` i vektor oznaka `y`) te vraća $(n+1)$-dimenzijski vektor težina tipa `ndarray`. Ako je `trace=True`, funkcija dodatno vraća listu (ili matricu) vektora težina $\mathbf{w}^0,\mathbf{w}^1,\dots,\mathbf{w}^k$ generiranih kroz sve iteracije optimizacije, od 0 do $k$. Optimizaciju treba provoditi dok se ne dosegne `max_iter` iteracija, ili kada razlika u pogrešci unakrsne entropije između dviju iteracija padne ispod vrijednosti `epsilon`. Parametar `alpha` predstavlja faktor L2-regularizacije.

Preporučamo definiranje pomoćne funkcije `lr_h(x,w)` koja daje predikciju za primjer `x` uz zadane težine `w`. Također, preporučamo i funkciju `cross_entropy_error(X,y,w)` koja izračunava pogrešku unakrsne entropije modela na označenom skupu `(X,y)` uz te iste težine.

**NB:** Obratite pozornost na to da je način kako su definirane oznake ($\{+1,-1\}$ ili $\{1,0\}$) kompatibilan s izračunom funkcije gubitka u optimizacijskome algoritmu.

```python
from numpy import linalg
from sklearn.metrics import log_loss

def lr_h(x, w): 
    feat = sklearn.preprocessing.PolynomialFeatures(1)
    X = feat.fit_transform(x)
    return sigmoid(np.dot(X, w))

def cross_entropy_error(X, y, w): 
    # Vaš kôd ovdje
    return log_loss(y, lr_h(X, w))

def lr_train(X, y, eta=0.01, max_iter=2000, trace=False, alpha=0, epsilon=0.000001):
    # Vaš kôd ovdje
    current_error = np.inf
    previous_error = 0
    tracing_matrix = []
    w = np.zeros(X.shape[1] + 1)
    i = 0
    
    while i < max_iter and abs(previous_error - current_error) > epsilon:
        
        delta_w = np.zeros(X.shape[1] + 1)
        h = lr_h(X, w)
        
        for j in range(X.shape[0]):
            delta_w[0] -= (h[j] - y[j])
            delta_w[1:] -= (h[j] - y[j])*X[j]
            
        w[0] += eta*delta_w[0]
        w[1:] = w[1:] * (1 - eta*alpha) + eta*delta_w[1:]
        
        tracing_matrix.append(np.copy(w))
        i += 1
        previous_error = current_error
        current_error = cross_entropy_error(X, y, w)
        # print(current_error)
        
    if trace:
        return (w, tracing_matrix)
    else:
        return w
```

Prikažite na jednom grafikonu pogrešku unakrsne entropije (očekivanje logističkog gubitka) i pogrešku klasifikacije (očekivanje gubitka 0-1) na skupu `seven` kroz iteracije optimizacijskog postupka.
```python
# Vaš kôd ovdje
from sklearn.metrics import zero_one_loss

W, trace_matrix = lr_train(seven_X, seven_y, trace=True)

cross_entropies = []
zero_ones = []

for w in trace_matrix:
    cross_entropies.append(cross_entropy_error(seven_X, seven_y, w))
    zero_ones.append(zero_one_loss(seven_y, lr_h(seven_X, w) > 0.5))

plt.plot(range(len(trace_matrix)), cross_entropies, label="Cross entropy loss")
plt.plot(range(len(trace_matrix)), zero_ones, label="Zero one loss")

etas = [0.005, 0.01, 0.05, 0.1]
plt.figure()
for eta in etas:
    cr_entropies = []
    W, trace_mat = lr_train(seven_X, seven_y, eta=eta, trace=True)
    for w in trace_mat:
        cr_entropies.append(cross_entropy_error(seven_X, seven_y, w))
    plt.plot(range(len(trace_mat)), cr_entropies, label="eta={0}".format(eta))
plt.legend(loc="best")
    
```
